## 17)Display the name of person having the highest basepay.

```python
data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')

# Now find the rows where 'BasePay' is equal to the maximum
data[data['BasePay'] == data['BasePay'].max()]
```

## 18.) Find average basepay of all employees per year.

```python
# Convert 'BasePay' to numeric, forcing errors to NaN
data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')

# Now you can group by 'Year' and calculate the mean of 'BasePay'
data.groupby('Year')['BasePay'].mean()
```

## 19) Find AverageBasePay of all employees per job title.

```python

# Convert 'BasePay' to numeric, forcing errors to NaN
data['BasePay'] = pd.to_numeric(data['BasePay'], errors='coerce')

# Group by 'JobTitle' and calculate the mean of 'BasePay'
data.groupby('JobTitle')['BasePay'].mean()
```
