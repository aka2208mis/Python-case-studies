## 20. Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
data.columns
```

```python
# Define thresholds for classification
def classify_rating(rating):
    if rating >= 8.0:
        return "Excellent"
    elif 6.0 <= rating < 8.0:
        return "Good"
    else:
        return "Average"

# Apply the classification to the 'Rating' column
data['Rating Category'] = data['Rating'].apply(classify_rating)

# Display the first few rows to verify
print(data[['Title', 'Rating', 'Rating Category']].head())
```

## 22. Find Unique Values From Genre

```python
data.columns
```

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
  for item1 in item:
    one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
  if item not in uni_list:
    uni_list.append(item)
```

```python
uni_list
```

```python
len(uni_list)
```
