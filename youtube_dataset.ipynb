## 6. Data Cleaning  (Replace '--'  to NaN)

```python
data.head(20)
```

```python
data['Subscribers']=data['Subscribers'].replace('--',np.nan,regex=True)
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.isnull())
```

```python
data.dropna(axis=0,inplace=True)
```

```python
data.isnull().sum()
```

## 8. Data Cleaning [ Rank Column ]

```python
data.head(5)
```

```python
data.tail(5)
```

```python
data.dtypes
```

```python
data['Rank']=data['Rank'].str[0:-2]
```

```python
data.head(5)
```

```python
data['Rank']=data['Rank'].str.replace(',','').astype('int')
```

```python
data.tail(5)
```

```python
data.dtypes
```

## 9. Data Cleaning [ Video Uploads & Subscribers ]

```python
data.head(6)
```

```python
data.dtypes
```

```python
data['Video Uploads'] = pd.to_numeric(data['Video Uploads'], errors='coerce').astype('Int64')

```

```python
data['Subscribers']=data['Subscribers'].astype('int')
```

```python
data.dtypes
```

## 10. Data Cleaning [ Grade Column ]

```python
data.head(5)
```

```python
data['Grade'].unique()
```

```python
data['Grade']=data['Grade'].map({'A++ ':5, 'A+ ':4, 'A ':3, 'A- ':2, 'B+ ':1, '\xa0 ':0})
```

```python
data.head(5)
```
