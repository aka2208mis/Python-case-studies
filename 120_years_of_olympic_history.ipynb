## Find and display the count of medals(Bronze, Silver, Gold) won in each city,and present the results in a table format.

## Sort the medal counts per city based on the total number of medals.

```python
data.head(1)
```

```python
medal_count = data.groupby(['City','Medal']).size().unstack()
```

```python
medal_count['Total']=medal_count[['Bronze','Gold','Silver']].sum(axis=1)
```

```python
medal_count
```

```python
medal_count.sort_values(by='Total',ascending=False)
```
